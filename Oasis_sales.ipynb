```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Advertising.csv')
```

```python
df.head()
```

we have to scale the data and build a regression model

```python
df=df.iloc[:,1:]
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
df = pd.DataFrame(df)
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df)
sdf = pd.DataFrame(scaled_data)
```

```python
sdf.head()
```

```python
x=sdf[[0,1,2]]
y=sdf[3]
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_ts,y_tr,y_ts=train_test_split(x,y,test_size=0.2)
```

lr-reg

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_tr,y_tr)
y_pred=lr.predict(x_ts)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print(mean_absolute_error(y_ts,y_pred))
print(mean_squared_error(y_ts,y_pred))
print(r2_score(y_ts,y_pred))
```

using rf

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_tr,y_tr)
y_pred1=rf.predict(x_ts)
```

```python
print(mean_absolute_error(y_ts,y_pred1))
print(mean_squared_error(y_ts,y_pred1))
print(r2_score(y_ts,y_pred1))
```

got a good accuracy of 98% using rf
